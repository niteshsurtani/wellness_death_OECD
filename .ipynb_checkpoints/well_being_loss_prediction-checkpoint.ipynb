## DATA PREPARATION

```python
directory = "data/"
df1 = pd.read_csv(directory + "delos_2013.csv")
df2 = pd.read_csv(directory + "delos_2014.csv")
df3 = pd.read_csv(directory + "delos_2015.csv")
df4 = pd.read_csv(directory + "delos_2016.csv")
```

```python

t1 = np.array([2014] * df2.shape[0])
df2['TIME'] = pd.Series(t1, index=df2.index)

t2 = np.array([2015] * df3.shape[0])
df3['TIME'] = pd.Series(t2, index=df3.index)

t3 = np.array([2016] * df4.shape[0])
df4['TIME'] = pd.Series(t3, index=df4.index)
```

```python
indicator_mapping = df1.loc[:,['INDICATOR','Indicator']]
indicator_mapping.drop_duplicates(inplace=True)
indicator_mapping
```

```python
df1 = df1[['Country', 'INDICATOR', 'TIME', 'Value']]
df1.drop_duplicates(inplace=True)

df2 = df2[['Country', 'INDICATOR', 'TIME', 'Value']]
df2.drop_duplicates(inplace=True)

df3 = df3[['Country', 'INDICATOR', 'TIME', 'Value']]
df3.drop_duplicates(inplace=True)

df4 = df4[['Country', 'INDICATOR', 'TIME', 'Value']]
df4.drop_duplicates(inplace=True)

frame = [df1, df2, df3, df4]
df = pd.concat(frame)

# Aggregated 3 years data
input_df = df.rename(index=str, columns={"TIME": "Year"})
```

## Grid Search with PCA and Ridge alpha paramaters

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

## K-Fold Cross Validation

pipe = make_pipeline(StandardScaler(), PCA(), StandardScaler(), Ridge())

param_grid = {'ridge__alpha': [0.001, 0.01, 0.1, 1, 10, 100], 'pca__n_components': range(2,11)}
grid = GridSearchCV(pipe, param_grid, cv=10, scoring='r2')
grid.fit(X_train, y_train)
print("Best estimator:\n{}".format(grid.best_estimator_))
```

### Best Params : PCA dim = 7, alpha = 0.001

```python
pipe = make_pipeline(StandardScaler(), PCA(n_components=7), StandardScaler(), Ridge(alpha=0.001))

ridge01 = pipe.fit(X_train, y_train)
print("Training set R2 score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set R2 score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
### 95% of information can be represented by 1 feature.
pca = PCA(n_components=7)
pca.fit(X_train)
print(pca.explained_variance_ratio_)
```

```python
def summ(X, y, predictions, params):
    newX = np.append(np.ones((len(X),1)), X, axis=1)
    print(newX.shape)
    MSE = (sum((y-predictions)**2))/(newX.shape[0]-newX.shape[1])

    # Note if you don't want to use a DataFrame replace the two lines above with
    # newX = np.append(np.ones((len(X),1)), X, axis=1)
    # MSE = (sum((y-predictions)**2))/(len(newX)-len(newX[0]))

    var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
    sd_b = np.sqrt(var_b)
    ts_b = params/ sd_b

    p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-1))) for i in ts_b]

    sd_b = np.round(sd_b,3)
    ts_b = np.round(ts_b,3)
    p_values = np.round(p_values,3)
    params = np.round(params,4)

    myDF3 = pd.DataFrame()
    myDF3["Coefficients"],myDF3["Standard Errors"],myDF3["t-values"],myDF3["p-values"] = [params,sd_b,ts_b,p_values]
    return myDF3
```
